```python
import collections

def flatten(d, parent_key='', sep='_'):
    items = []
    for k, v in d.items():
        new_key = parent_key + sep + k if parent_key else k
        if isinstance(v, collections.MutableMapping):
            items.extend(flatten(v, new_key, sep=sep).items())
        else:
            items.append((new_key, v))
    return dict(items)

print(flatten(new_dict))
```

<h3>Fantasy Data</h3>

```python
# Identify current offensive team
pbp_df['offteam'] = np.where(pbp_df['defteam']==pbp_df['home_team'], 
                             pbp_df['away_team'], 
                             pbp_df['home_team'])

# Calculate yards the ball moves forward from current play to next play
pbp_df['next_yardline_100'] = pbp_df['yardline_100'].shift(-1)
pbp_df['yards_forward'] = pbp_df['yardline_100'] - pbp_df['next_yardline_100']
pbp_df.drop(columns=['next_yardline_100'])

# Calculate Receiving fantasy points 
pbp_df['fantasy_rec_yd_pts'] = (pbp_df['receiving_yards']*pt_rec_yd) + pt_rec
pbp_df['fantasy_rec_td_pts'] = np.where((pbp_df['td_player_id'].notnull())&(pbp_df['play_type']=='pass'),
                                        pt_rec_td,
                                        0)
pbp_df['fantasy_rec_twoPointConv_pts'] = np.where((pbp_df['two_point_conv_result']=='success')&
                                                  (pbp_df['play_type']=='pass'),
                                                  pt_twoPointConv,
                                                  0)
pbp_df['fantasy_rec_pts'] = pbp_df['fantasy_rec_yd_pts'] + \
                            pbp_df['fantasy_rec_td_pts'] + \
                            pbp_df['fantasy_rec_twoPointConv_pts']
                                                
# Calculate Rushing fantasy points 
pbp_df['fantasy_rush_yd_pts'] = pbp_df['rushing_yards']*pt_rush_yd
pbp_df['fantasy_rush_td_pts'] = np.where((pbp_df['td_player_id'].notnull())&(pbp_df['play_type']=='run'),
                                         pt_rush_td,
                                         0)
pbp_df['fantasy_rush_lostFumble_pts'] = np.where(pbp_df['fumble_lost']==1,
                                                 pt_rush_lostFumble,
                                                 0)
pbp_df['fantasy_rush_twoPointConv_pts'] = np.where((pbp_df['two_point_conv_result']=='success')&
                                                   (pbp_df['play_type']=='run'),
                                                   pt_twoPointConv,
                                                   0)
pbp_df['fantasy_rush_pts'] = pbp_df['fantasy_rush_yd_pts'] + \
                             pbp_df['fantasy_rush_td_pts'] + \
                             pbp_df['fantasy_rush_lostFumble_pts'] + \
                             pbp_df['fantasy_rush_twoPointConv_pts']
            
# Calculate Passing fantasy points
pbp_df['fantasy_pass_yd_pts'] = pbp_df['passing_yards']*pt_pass_yd
pbp_df['fantasy_pass_td_pts'] = np.where((pbp_df['td_player_id'].notnull())&(pbp_df['play_type']=='pass'),
                                         pt_pass_td,
                                         0)
pbp_df['fantasy_pass_int_pts'] = np.where(pbp_df['interception']==1,
                                          pt_pass_int,
                                          0)
pbp_df['fantasy_pass_twoPointConv_pts'] = np.where((pbp_df['two_point_conv_result']=='success')&
                                                   (pbp_df['play_type']=='pass'),
                                                   pt_twoPointConv,
                                                   0)
pbp_df['fantasy_pass_pts'] = pbp_df['fantasy_pass_yd_pts'] + \
                             pbp_df['fantasy_pass_td_pts'] + \
                             pbp_df['fantasy_pass_int_pts'] + \
                             pbp_df['fantasy_pass_twoPointConv_pts']
```

```python
rush_df = off_pbp_df.merge(off_players_df, 
                           how='inner', 
                           left_on='rusher_id', 
                           right_on='player_id') \
                    .groupby(column_list)['fantasy_rush_pts'].sum() \
                    .reset_index() \
                    .sort_values(by=['fantasy_rush_pts'], ascending=False)
rec_df = off_pbp_df.merge(off_players_df, 
                          how='inner', 
                          left_on='receiver_id', 
                          right_on='player_id') \
                   .groupby(column_list)['fantasy_rec_pts'].sum() \
                   .reset_index() \
                   .sort_values(by=['fantasy_rec_pts'], ascending=False)
pass_df = off_pbp_df.merge(off_players_df, 
                           how='inner', 
                           left_on='passer_id', 
                           right_on='player_id') \
                    .groupby(column_list)['fantasy_pass_pts'].sum() \
                    .reset_index() \
                    .sort_values(by=['fantasy_pass_pts'], ascending=False)
off_df = rush_df.merge(rec_df[['player_id','fantasy_rec_pts']],
                       how='left',
                       on='player_id') \
                .merge(pass_df[['player_id','fantasy_pass_pts']],
                       how='left',
                       on='player_id') \
                .fillna(0)
off_df['fantasy_total_pts'] = off_df['fantasy_rush_pts'] + \
                              off_df['fantasy_rec_pts'] + \
                              off_df['fantasy_pass_pts']
off_df.sort_values(by=['fantasy_total_pts'], ascending=False)
```

```python
off_df['position_rank'] = off_df.groupby('position')['fantasy_total_pts'] \
                                .rank(method='dense', ascending=False) \
                                .astype(int)
off_df['keep'] = np.where(off_df['position'].isin(['WR','RB']),
                          np.where(off_df['position_rank'] <= 25, 1, 0),
                          np.where(off_df['position_rank'] <= 25, 1, 0))
top_off_df = off_df[(off_df['keep']==1)&(off_df['position']=='WR')]
```
